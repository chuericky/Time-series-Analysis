### This program simulates and visualizes the signal-plus-noise model.
Source: Time Series Analysis and Its Applications, 4th Edition, Shumway and Stoffer (2016), Exercise 1.2.

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

%matplotlib inline
```

Consider a signal-plus-noise model of the general form $x_t = s_t + w_t$ , where $w_t$ is Gaussian white noise with $\sigma_w^2 = 1$. Simulate and plot $n = 200$ observations from each of the following two models.

a. Model 1:
- $x_t = s_t + w_t$, for $t = 1, ..., 200$, where
$\begin{cases}
  s_t = 0, &t = 1, ..., 100\\    
  s_t = 10 \exp\{-\frac{t - 100}{20}\}\cos(2\pi t/4),  &t = 101, ..., 200  
\end{cases}$

```python
t = np.arange(1,201,1)
t_front = np.arange(1,101,1)
t_back = np.arange(101,201,1)

### Signal
st1_front = np.array([0.] * 100)
st1_back = 10 * np.exp(-(t_back-100.)/20.) * np.cos(np.pi * t_back / 2)
st1 = np.append(st1_front, st1_back)

### White noise
wt1 = np.random.normal(0, 1, 200)

### Signal-plus-noise
xt1 = st1 + wt1
```

```python
plt.figure(figsize = (7,5))
plt.xlabel(r"$t$", fontsize = 14)
plt.ylabel(r"$x_t = s_t + w_t$", fontsize = 14)
plt.title("Model 1", fontsize = 16)
plt.plot(t, xt1)
```

b. Model 2:
- $x_t = s_t + w_t$, for $t = 1, ..., 200$, where
$\begin{cases}
  s_t = 0, &t = 1, ..., 100\\    
  s_t = 10 \exp\{-\frac{t - 100}{200}\}\cos(2\pi t/4),  &t = 101, ..., 200  
\end{cases}$

```python
### Signal
st2_front = np.array([0.] * 100)
st2_back = 10 * np.exp(-(t_back-100.)/200.) * np.cos(np.pi * t_back / 2)
st2 = np.append(st2_front, st2_back)

### White noise
wt2 = np.random.normal(0, 1, 200)

### Signal-plus-noise
xt2 = st2 + wt2
```

```python
plt.figure(figsize = (7,5))
plt.xlabel(r"$t$", fontsize = 14)
plt.ylabel(r"$x_t = s_t + w_t$", fontsize = 14)
plt.title("Model 2", fontsize = 16)
plt.plot(t, xt2)
```

c. Model 1 in part (a) has similar appearance with the explosion series shown in Fig. 1.7. in the book, that there is a large fluctuation (noise) immediately after $t = 100$, but the plot decays as time progresses.

Whereas model 2 resembles the earthquake pattern. The flucuation is large after $t = 100$, and the decay time is much longer than that of model 1, i.e. the large fluctuation lasts for a longer time.

For $t = 1, 2, ..., 100$, plot of the signal modulator (a) $exp\{-t/20\}$ and (b) $exp\{-t/200\}$.

```python
sigmod_a = np.exp(-t_front/20.)
sigmod_b = np.exp(-t_front/200.)

plt.figure(figsize = [7,5])
plt.plot(t_front, sigmod_a)
plt.plot(t_front, sigmod_b)
plt.legend([r"$\exp(-t/20)$", r"$\exp(-t/200)$"], fontsize = 14)
plt.title(r"Signal modulators for $t = 1,2,...,100$", fontsize = 16)
plt.xlabel(r"Time, $t$", fontsize = 14)
plt.ylabel("Signal modulator", fontsize = 14)
```

The decay time of modulator $\exp(-t/20)$ is much shorter than that of $\exp(-t/200)$.  The modulator $\exp(-t/20)$ is close to 0 when $t \sim 100$.
